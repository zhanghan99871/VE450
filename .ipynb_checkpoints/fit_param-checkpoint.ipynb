```python
dataset = [
    ('data/VW216/VW216.RTSL-BUL.HV_6809.47.png', 6809.47),
    ('data/VW310/VW310-6CS.DRL-20220328.HV_2654.74.png', 2654.74),
    ('data/VW310-PL/VW310-6CS.PL-FTSL-20220401.HV_1744.43.png', 1744.43),
    ('data/VW316/VW316 7CS.RTSL-BUL-SL-TL.HV_2124.45.png', 2124.45),
    ('data/VW323/VW323 0CS.SL-RTSL-BUL-RFL.HV_2381.67.png', 2381.67),
    ('data/VW326/VW326 0CS.SL-TL-RTSL-BUL-RFL.HV_9001.23.png', 9001.23),
    ('data/VW331/VW331_Basic_CHL_simulation setting.DRL_PL_FTSL_20220311.HV_15241.2.png', 15241.2),
    ('data/VW316-TLB/VW316 7CS-RCL.TLB-20220810.HV_25.1441.png', 25.1441),
    ('data/VW323-TL/VW323 0CS.TL.HV_49.3145.png', 49.3145)
]

mean_lab_luminance = []
real_luminance = []

for png_file, initial_luminance in dataset:
    image = RawImage()
    image.loadRGB(png_file)
    image.convert_rgb_to_lab_luminance()
    mean_lab_luminance.append(np.mean(image.luminance[image.luminance > 0]))
    real_luminance.append(initial_luminance)

plt.scatter(mean_lab_luminance, real_luminance)
plt.xlabel('mean lab luminance')
plt.ylabel('real initial luminance');
```

```python
original_img = cv.imread('data/VW216/VW216.RTSL-BUL.HV_6809.47.png')
lab_original = cv.cvtColor(original_img, cv.COLOR_BGR2LAB)
l, a, b = cv.split(lab_original)
l, a, b = l.ravel(), a.ravel(), b.ravel()
#l_sample, a_sample, b_sample = np.random.choice(l.ravel(), 10000), np.random.choice(a.ravel(), 10000), np.random.choice(b_ch.ravel(), 10000)
X_a = pd.DataFrame({'L': l, 'a': a, 'a^2': a**2})
X_a = sm.add_constant(X_a)
X_b = pd.DataFrame({'L': l, 'b': b, 'b^2': b**2})
X_b = sm.add_constant(X_b)

intercept_a = []
param_L_a = []
param_a = []
param_a2 = []
intercept_b = []
param_L_b = []
param_b = []
param_b2 = []

for i in range(1, 21):   
    processed_img = cv.imread(f'data/VW216/VW216.RTSL-BUL.HV_00{i}.png')
    lab_processed = cv.cvtColor(processed_img, cv.COLOR_BGR2LAB)
    l1, a1, b1 = cv.split(lab_processed)
    l1, a1, b1 = l1.ravel(), a1.ravel(), b1.ravel()

    y_a = pd.DataFrame({'y_a': a1})
    y_b = pd.DataFrame({'y_b': b1})
    
    model_a = sm.OLS(y_a, X_a)
    result_a = model_a.fit()
    params_a = result_a.params.to_numpy()
    intercept_a.append(params_a[0])
    param_L_a.append(params_a[1])
    param_a.append(params_a[2])
    param_a2.append(params_a[3])

    model_b = sm.OLS(y_b, X_b)
    result_b = model_b.fit()
    params_b = result_b.params.to_numpy()
    intercept_b.append(params_b[0])
    param_L_b.append(params_b[1])
    param_b.append(params_b[2])
    param_b2.append(params_b[3])

index = range(1, 21)

plt.figure(figsize=(20, 8))

plt.subplot(2,4,1)
plt.scatter(index, intercept_a)
plt.xlabel('index')
plt.ylabel('intercept_a')

plt.subplot(2,4,2)
plt.scatter(index, param_L_a)
plt.xlabel('index')
plt.ylabel('param_L_a')

plt.subplot(2,4,3)
plt.scatter(index, param_a)
plt.xlabel('index')
plt.ylabel('param_a')

plt.subplot(2,4,4)
plt.scatter(index, param_a2)
plt.xlabel('index')
plt.ylabel('param_a^2')

plt.subplot(2,4,5)
plt.scatter(index, intercept_b)
plt.xlabel('index')
plt.ylabel('intercept_b')

plt.subplot(2,4,6)
plt.scatter(index, param_L_b)
plt.xlabel('index')
plt.ylabel('param_L_b')

plt.subplot(2,4,7)
plt.scatter(index, param_b)
plt.xlabel('index')
plt.ylabel('param_b')

plt.subplot(2,4,8)
plt.scatter(index, param_b2)
plt.xlabel('index')
plt.ylabel('param_b^2');
```

```python
X_a = pd.DataFrame({'L': l, 'a': a, 'a^2': a**2})
X_a = sm.add_constant(X_a)
y_a = pd.DataFrame({'y_a': a1})
X_b = pd.DataFrame({'L': l, 'b': b, 'b^2': b**2})
X_b = sm.add_constant(X_b)
y_b = pd.DataFrame({'y_b': b1})
X_a_train, X_a_test, y_a_train, y_a_test = train_test_split(X_a, y_a, test_size=0.2, random_state=42)
X_b_train, X_b_test, y_b_train, y_b_test = train_test_split(X_b, y_b, test_size=0.2, random_state=42)

model_a = sm.OLS(y_a_train, X_a_train)
result_a = model_a.fit()
print(result_a.summary())

model_b = sm.OLS(y_b_train, X_b_train)
result_b = model_b.fit()
print(result_b.summary())
```
